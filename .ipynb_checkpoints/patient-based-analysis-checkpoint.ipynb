```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("databaseForStudents.csv")
df.head(10)
```

### Creating a Dataframe with Patient Related Data

## 1.1 Gender Distribution For all Patients

```python
df_noname_dropped = df.dropna(subset=["name"])
df_Gender_data = df_noname_dropped[[ "name", "gender"]]
df_Gender_data.head(5)
```

```python
Gender_counts = df_Gender_data["gender"].value_counts()
print(Gender_counts)
```

```python
%matplotlib inline
plot = Gender_counts.plot(kind='pie', explode= [0.1, 0], shadow= True, autopct='%1.1f%%')
```
